```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
B_data = pd.read_csv("C:/Users/Yash Tandon/Desktop/python course/ML projects/US-border-crossing-EDA/Border_Crossing_Entry_Data.csv")
B_data.head()
```

```python
B_data.info()
```

```python
B_name = B_data['Border'].unique()
print(B_name)
```

```python
print(B_data.groupby('Border').size())
sns.countplot(B_data['Border'], label = 'count')
plt.show()
```

```python
print(B_data.groupby('Measure').size())
plt.figure(figsize = (30, 10))
sns.countplot(B_data['Measure'], label = 'Measure')
plt.show()
```

```python
B_data['Date'] = pd.to_datetime(B_data['Date'])
```

```python
B_year = B_data['Date'].map(lambda x : x.year).unique()
print(B_year)
```

```python
print(B_data.groupby(B_data['Date'].map(lambda x : x.year)).size())
plt.figure(figsize = (20,5))
sns.countplot(B_data['Date'].map(lambda x : x.year), label = 'year')
plt.show()
```
